```python
cod_continents_countries = df_pandas["Region"].str.split('[-/]',expand=True) #Si al split ponemos [] adentro aceptara varios condiciones para separar
cod_continents_countries.columns = ['Cod_Pais', 'Continente','Pais']

df_pandas = pd.concat([df_pandas, cod_continents_countries], axis=1)
df_pandas
```

```python
df_pandas['Cod_Pais'] = df_pandas['Cod_Pais'].str.replace(" ","")
df_pandas['Continente'] = df_pandas['Continente'].str.replace(" ","")
df_pandas
```

```python
trimestre_year = df_pandas["Periodo"].str.split(expand=True) #Si al split ponemos [] adentro aceptara varios condiciones para separar
trimestre_year.columns = ['Trimestre', 'Anio']

df_pandas = pd.concat([df_pandas, trimestre_year], axis=1)
df_pandas
```

```python
df_pandas.drop(columns=['Region'], inplace=True)
df_pandas.drop(columns=['Periodo'], inplace=True)
```

```python
df_pandas['Trimestre'] = df_pandas['Trimestre'].str.replace('Q','T')
df_pandas
```

```python
df_pandas['Trimestre'].unique().tolist() #Para obtener datos como un Distinct
df_pandas['Anio'].unique().tolist() #Para obtener datos como un Distinct
df_pandas['Tipo Producto'].unique().tolist() #Para obtener datos como un Distinct
```
